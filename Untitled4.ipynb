# Text Preprocessing

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
```

```python
# Download necessary resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
# Initialize tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# Function for text preprocessing
def preprocess_text(text):
    text = text.lower()  # Lowercasing
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    tokens = word_tokenize(text)  # Tokenization
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]  # Remove stopwords & numbers
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatization
    return " ".join(tokens)
```

```python
# Apply preprocessing to premise and hypothesis columns
df['premise_clean'] = df['premise'].apply(preprocess_text)
df['hypothesis_clean'] = df['hypothesis'].apply(preprocess_text)

# Show some cleaned text
print(df[['premise_clean', 'hypothesis_clean']].head())
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd


# Ensure preprocessing is done
if 'premise_clean' not in df or 'hypothesis_clean' not in df:
    print("Preprocessing step missing. Run text preprocessing first.")
else:
    # Combine premise and hypothesis
    df['combined_text'] = df['premise_clean'] + " " + df['hypothesis_clean']

    # Initialize TF-IDF Vectorizer
    vectorizer = TfidfVectorizer(max_features=5000)  # Limit to top 5000 features
    X_tfidf = vectorizer.fit_transform(df['combined_text'])

    # Convert labels to numpy array
    y = df['label'].values

    print("TF-IDF Feature Extraction Complete. Shape:", X_tfidf.shape)

    # Optional: Show some feature names
    print("Sample TF-IDF features:", vectorizer.get_feature_names_out()[:10])
```

# Model training and evaluation

## Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Train Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluation
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```

## Using LSTM

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split

# Hyperparameters
MAX_WORDS = 5000  # Vocabulary size
MAX_LEN = 100     # Max length of each sequence
EMBEDDING_DIM = 100

# Tokenize text
tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
tokenizer.fit_on_texts(df['combined_text'])
X_seq = tokenizer.texts_to_sequences(df['combined_text'])
X_pad = pad_sequences(X_seq, maxlen=MAX_LEN, padding='post')

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_pad, y, test_size=0.2, random_state=42)

# Build LSTM model
model = Sequential([
    Embedding(MAX_WORDS, EMBEDDING_DIM, input_length=MAX_LEN),
    LSTM(128, return_sequences=True),
    LSTM(64),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')  # 3 classes
])

# Compile model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))

# Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
print("LSTM Model Accuracy:", accuracy)
```
